## Actual Plots

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import os
name_ = os.listdir()
names = [i for i in name_ if i[-3:]=='csv']

skip_data = 1 # 1 mane data ektao skip korbe na
# Koto point skip korte chaash, Joto barabi toto smooth
# eg: [1,2,3,4,5], skip_data = 2 mane [1,3,5] nebe sudhu, kom ghinji lagbe

for name in names:
  phase = name.split('_')[1] # Phase ta store rakhbe
  skip_rows=13
  file_path=f'/content/{name}'
  data=pd.read_csv(file_path,skiprows=skip_rows)
  plt.figure(figsize=(10,6))
  plt.scatter(data["TIME"][::skip_data],data['CH1'][::skip_data],label='CH1', marker = '*', s = 1)
  plt.scatter(data["TIME"][::skip_data],data['CH2'][::skip_data],label='CH2', marker = '*', s = 1)

  plt.xlabel('Time')
  plt.ylabel('Values')
  plt.title(f'Time vs CH1 and CH2 | phase: {phase} | filename: {name}') # Ja title dorkar ekhane
  plt.legend()
  plt.show()
```

## Smoothened Plots

```python
from scipy.signal import savgol_filter


import os
name_ = os.listdir()
names = [i for i in name_ if i[-3:]=='csv']

window = 200

for name in names:
  phase = name.split('_')[1] # Phase ta store rakhbe
  skip_rows=13
  file_path=f'/content/{name}'
  data=pd.read_csv(file_path,skiprows=skip_rows)
  plt.figure(figsize=(10,6))
  data['CH1_smoothed'] = savgol_filter(data['CH1'],
                                       window_length=window,
                                       polyorder=1)
  data['CH2_smoothed'] = savgol_filter(data['CH2'],
                                       window_length=4*window,
                                       polyorder=1)

  plt.figure(figsize=(10,6))
  plt.plot(data["TIME"],data['CH1_smoothed'],label='CH1_smoothened')
  plt.plot(data["TIME"],data['CH2_smoothed'],label='CH2_smoothened')

  plt.xlabel('Time')
  plt.ylabel('Values')
  plt.title(f'Smoothened Time vs CH1 and CH2 | phase: {phase} | filename: {name}')
  plt.legend()
  plt.show()
```
